```python
import requests
from bs4 import BeautifulSoup
import re
from selenium import webdriver
```

```python
driver_path = 'C:/Users/alisson.osanches/Downloads/chromedriver_win32/chromedriver'
driver = webdriver.Chrome(executable_path = driver_path)
url = 'https://www.infomoney.com.br/cryptos/cotacoes'
driver.get(url)
soup = BeautifulSoup(driver.page_source, 'html.parser')
driver.close()
```

```python
table = soup.find('table', attrs={'class':'display border-t-brown-3'})

table_body = table.find('tbody')
#print(table_body.prettify())
```

```python
data = []
table_body = table.find('tbody')
rows = table_body.find_all('tr')
for row in rows:
	cols = row.find_all('td')
	cols = [ele.text.strip() for ele in cols]
	data.append([ele for ele in cols if ele])

#print(data)
```

```python
def getCents(text, isFloat = 0):
    if(isFloat == 1):
        return float(''.join(re.findall('\d+\.\d+', text)))
    return int(''.join(re.findall('\d+', text)))
```

```python
data_insert = []

for line in data:
    item = {
        'coin_id': line[0], 
        'names': line[1], 
        'val_mercado': getCents(line[2]), 
        'price_real': getCents(line[3]), 
        'price_btc': line[4].split()[1], 
        'val_tot_day': getCents(line[5]), 
        'variation_day': line[6] 
    }
    data_insert.append(item)
    
print(data_insert)
```
